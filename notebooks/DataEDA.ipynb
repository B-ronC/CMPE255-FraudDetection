# EDA

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

### Target Variable Distribution

```python
# Visualize class imbalance
class_counts = df["Class"].value_counts().sort_index()  # 0 = non-fraud, 1 = fraud
labels = ["Legitimate = 0", "Fraud = 1"]

# Bar plot
plt.figure(figsize=(6, 4))
plt.bar(labels, class_counts)
plt.title("Class Distribution", pad=20)
plt.ylabel("Count")
for i, v in enumerate(class_counts):
    plt.text(i, v, str(v), ha="center", va="bottom")
plt.show()

# Pie chart
plt.figure(figsize=(6, 6))
plt.pie(
    class_counts,
    labels=labels,
    autopct="%1.3f%%",
    startangle=90
)
plt.title("Class Proportion", pad=20)
plt.axis("equal")
plt.tight_layout()
plt.show()
```

Large class imbalance

### Amount Distribution by Class

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="Class", y="Amount", data=df)
plt.title("Boxplot of Transaction Amount by Class")
plt.xlabel("Class (0 = Legitimate, 1 = Fraud)")
plt.ylabel("Transaction Amount")
plt.yscale("log") 
plt.show()
```

Fraudulent transactions tend to have lower amounts

### Fraud Frequency by Hour

```python
# Convert Time (seconds) → hour
df["Hour"] = (df["Time"] // 3600).astype(int) % 24

# Compute fraud counts for each hour
fraud_by_hour = (
    df[df["Class"] == 1]["Hour"]
    .value_counts()
    .reindex(range(24), fill_value=0)
)

# Plot fraud frequency by hour
plt.figure(figsize=(12, 5))
plt.bar(fraud_by_hour.index, fraud_by_hour.values, color="red", alpha=0.7)
plt.title("Fraud Frequency by Hour")
plt.xlabel("Hour of Day (0–23)")
plt.ylabel("Number of Fraudulent Transactions")
plt.xticks(range(24))
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.show()
```

Certain hours are riskier than others

### Fraud Correlation Analysis

```python
# Select only V1–V28 + Class
features = [f"V{i}" for i in range(1, 29)]
corrs = df[features + ["Class"]].corr()["Class"].drop("Class")

# Top 5 positive & negative correlations
top5_pos = corrs.sort_values(ascending=False).head(5)
top5_neg = corrs.sort_values().head(5)

print("Top 5 positive correlations:\n", top5_pos)
print("\nTop 5 negative correlations:\n", top5_neg)

# Bar plots
plt.figure(figsize=(12,5))
plt.subplot(1, 2, 1)
top5_pos.plot(kind="bar", color="red")
plt.title("Top 5 Positive Correlations with Fraud")
plt.ylabel("Correlation")

plt.subplot(1, 2, 2)
top5_neg.plot(kind="bar", color="green")
plt.title("Top 5 Negative Correlations with Fraud")
plt.ylabel("Correlation")

plt.tight_layout()
plt.show()
```
